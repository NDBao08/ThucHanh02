```python
import pandas as pd
df = pd.read_csv(r"C:\Users\trvum\Downloads\playground-series-s5e4\test.csv")
df.head(10)
```

```python
# Lọc các dòng có Publication_Day là "Monday"
df_monday = df[df["Publication_Day"] == "Monday"]

# Số lượng NaN theo từng cột
nan_per_column = df_monday.isnull().sum()
print(nan_per_column)

# Nội suy các giá trị số bị thiếu theo phương pháp tuyến tính
df_monday_interpolated = df_monday.interpolate(method='linear')

# Kiểm tra lại xem còn giá trị NaN không
print(df_monday_interpolated.isnull().sum())

# Xóa các dòng còn chứa giá trị thiếu (NaN) sau khi nội suy
monday_df_cleaned = df_monday_interpolated.dropna()

# Kiểm tra lại số giá trị thiếu sau khi xóa
print(monday_df_cleaned.isna().sum())
print(f"Số dòng còn lại sau khi xóa: {len(monday_df_cleaned)}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.structural import UnobservedComponents

# Đảm bảo cột thời gian (nếu có), hoặc dùng index giả định
monday_df_cleaned.reset_index(drop=True, inplace=True)

# Khởi tạo mô hình Local Level (level-only model)
model = UnobservedComponents(
    endog=monday_df_cleaned['Episode_Length_minutes'],
    level='local level'
)

# Fit mô hình với Kalman Filter
results = model.fit()

# Tóm tắt kết quả
print(results.summary())

# Dự báo (forecast) tiếp 10 bước
forecast = results.get_forecast(steps=10)
forecast_mean = forecast.predicted_mean
conf_int = forecast.conf_int()

# Vẽ kết quả
plt.figure(figsize=(12, 6))
plt.plot(monday_df_cleaned['Episode_Length_minutes'], label='Observed')
plt.plot(forecast_mean.index, forecast_mean, label='Forecast', color='red')
plt.fill_between(forecast_mean.index,
                 conf_int.iloc[:, 0],
                 conf_int.iloc[:, 1], color='pink', alpha=0.3)
plt.legend()
plt.title("Local Level Model Forecast")
plt.xlabel("Time Index")
plt.ylabel("Episode_Length_minutes")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.structural import UnobservedComponents

# Đặt lại index để dùng như trục thời gian
monday_df_cleaned.reset_index(drop=True, inplace=True)

# Khởi tạo mô hình xu hướng tuyến tính cục bộ
model = UnobservedComponents(
    endog=monday_df_cleaned['Episode_Length_minutes'],
    level='local linear trend'
)

# Huấn luyện mô hình
results = model.fit()

# In tóm tắt kết quả
print(results.summary())

# Dự báo 10 bước tiếp theo
forecast = results.get_forecast(steps=10)
forecast_mean = forecast.predicted_mean
conf_int = forecast.conf_int()

# Vẽ biểu đồ
plt.figure(figsize=(12, 6))
plt.plot(monday_df_cleaned['Episode_Length_minutes'], label='Observed')
plt.plot(forecast_mean.index, forecast_mean, label='Forecast', color='red')
plt.fill_between(forecast_mean.index, conf_int.iloc[:, 0], conf_int.iloc[:, 1],
                 color='pink', alpha=0.3)
plt.title("Local Linear Trend Model Forecast (Kalman Filter)")
plt.xlabel("Time Index")
plt.ylabel("Episode_Length_minutes")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.structural import UnobservedComponents

# Đặt lại index để giả định là thời gian
monday_df_cleaned.reset_index(drop=True, inplace=True)

# Khởi tạo mô hình xu hướng + chu kỳ (seasonal)
model = UnobservedComponents(
    endog=monday_df_cleaned['Episode_Length_minutes'],
    level='local linear trend',
    seasonal=7  # giả định có chu kỳ hàng tuần (7 ngày)
)

# Huấn luyện mô hình với Kalman Filter
results = model.fit()

# In tóm tắt kết quả
print(results.summary())

# Dự báo 14 bước tiếp theo
forecast = results.get_forecast(steps=14)
forecast_mean = forecast.predicted_mean
conf_int = forecast.conf_int()

# Vẽ kết quả
plt.figure(figsize=(12, 6))
plt.plot(monday_df_cleaned['Episode_Length_minutes'], label='Observed')
plt.plot(forecast_mean.index, forecast_mean, label='Forecast', color='red')
plt.fill_between(forecast_mean.index,
                 conf_int.iloc[:, 0], conf_int.iloc[:, 1],
                 color='pink', alpha=0.3)
plt.title("Seasonal + Trend Model Forecast (Kalman Filter)")
plt.xlabel("Time Index")
plt.ylabel("Episode_Length_minutes")
plt.legend()
plt.tight_layout()
plt.show()
```
